Preprocess the training data (audio of playing scale on a violin). This oould have been done inside the Training part, but this can be a tutorial of the audio tools at hand so I made this as its own notebook.

Conduct STFT (short-time Fourier transform) and save it to .npy file.

```python
import utils
import librosa
import IPython.display as ipd
import numpy as np
```

```python
audio, sr = librosa.load("./data/training/Scale.m4a")
```

```python
stft = utils.audio2stft(audio, sr)
stft = stft[:,85:-65] # clip to remove silent parts.
print(f"stft.shape:{stft.shape}")
utils.plot_stft(stft, sr)
```

```python
reconstructed_audio = utils.stft2audio(stft)
ipd.Audio(reconstructed_audio, rate=sr)
```

```python
np.save("./data/training/Scale.npy", stft)
```
